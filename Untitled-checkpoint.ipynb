```python
import yaml
import pandas as pd
import os
import glob
import numpy as np

# Define your data folders based on your desktop path structure
folders = {
    'T20': '../Data/t20s_male/',
    'ODI': '../Data/odis_male/',
    'Test': '../Data/tests_male/'
}

all_deliveries = []
print("Starting YAML parsing engine... Please wait a moment.")

# Loop through each format folder
for format_name, folder_path in folders.items():
    if not os.path.exists(folder_path):
        continue
        
    # Find all .yaml files in that folder
    yaml_files = glob.glob(os.path.join(folder_path, "*.yaml"))
    print(f"Found {len(yaml_files)} files for {format_name}. Processing the first 30 files...")
    
    # Process the first 30 files to create our clean testing database
    for file_path in yaml_files[:30]: 
        match_id = os.path.basename(file_path).replace('.yaml', '')
        
        try:
            with open(file_path, 'r') as file:
                match_data = yaml.safe_load(file)
            
            # Navigate into the Cricsheet innings structure
            if 'innings' in match_data:
                for inning in match_data['innings']:
                    for inning_name, inning_info in inning.items():
                        if 'deliveries' in inning_info:
                            for delivery in inning_info['deliveries']:
                                for ball_num, ball_data in delivery.items():
                                    
                                    # Extract ball-by-ball rows safely
                                    delivery_row = {
                                        'match_id': match_id,
                                        'format': format_name,
                                        'over': float(ball_num),
                                        'striker': ball_data.get('batsman'),
                                        'bowler': ball_data.get('bowler'),
                                        'runs_off_bat': ball_data.get('runs', {}).get('off_bat', 0),
                                        'extras': ball_data.get('runs', {}).get('extras', 0),
                                        'wicket_type': ball_data.get('wicket', {}).get('kind', None),
                                        'player_dismissed': ball_data.get('wicket', {}).get('player_out', None)
                                    }
                                    all_deliveries.append(delivery_row)
        except Exception as e:
            continue

# Convert our list of data into a structured Pandas DataFrame
master_df = pd.DataFrame(all_deliveries)
print(f"Success! Extracted {master_df.shape[0]} individual ball rows.")
master_df.head()
```

```python
# Function to automatically tag T20/ODI match phases
def assign_match_phase(over):
    if over < 6.0: return 'Powerplay'
    elif over < 15.0: return 'Middle Overs'
    else: return 'Death Overs'

# Apply our metrics engine logic
master_df['match_phase'] = master_df['over'].apply(assign_match_phase)
master_df['is_four'] = np.where(master_df['runs_off_bat'] == 4, 1, 0)
master_df['is_six'] = np.where(master_df['runs_off_bat'] == 6, 1, 0)
master_df['is_dot'] = np.where((master_df['runs_off_bat'] == 0) & (master_df['extras'] == 0), 1, 0)

# Save this output directly into your Data folder as a master CSV for SQL!
output_path = "../Data/master_cricket_cleaned.csv"
master_df.to_csv(output_path, index=False)

print(f"Day 2 Complete! Your SQL-ready file is saved at: {output_path}")
master_df[['format', 'over', 'match_phase', 'striker', 'runs_off_bat', 'is_dot']].head()
```
